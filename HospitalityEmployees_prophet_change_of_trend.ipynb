```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
%matplotlib inline
from prophet.plot import add_changepoints_to_plot
```

```python
df = pd.read_csv(r'C:\Users\nico_\Desktop\Time series\UDEMY_TSA_FINAL\UDEMY_TSA_FINAL\Data\HospitalityEmployees.csv')
```

```python
df.columns = ['ds','y']
```

```python
df['ds'] = pd.to_datetime(df['ds'])
```

```python
df.plot(x='ds', y='y', figsize=(10,5))
```

```python
m = Prophet()
m.fit(df)
future = m.make_future_dataframe(periods=12 ,freq='MS')
forecast = m.predict(future)
```

# Afficher les points de rupture (changement de tendance)

fig.gca() rcupre laxe principal de la figure (gca = get current axis)

```python
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(),m,forecast)
```
